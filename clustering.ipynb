```python
import pandas as pd

data = pd.read_csv('dataframe/champions_23_24.csv')

```

```python
# Seleccionar el último partido jugado por cada equipo local
data = data.groupby('Local').last().reset_index()

# Mostrar el resultado
data
```

```python
data.info()
```

```python
features = data[['GolesLocal', '%_Equipo1_Ganado', '%_1_G_Temporada',
                '%_1_G_Temporada_L', '%_1_E_Temporada_L', '%_1_P_Temporada_L', '1_Media_G',
                '1_Media_G_Local', '1_Media_Goles_PP', '1_ValorJugadores', '1_MediaJugadores']]


features.sample(10)
```

Hemos seleccionado las características para cada instancia de un equipo. Dado que tenemos 12 características, podríamos interpretarlo como coordenadas que describen una instancia en un espacio de doce diemnsiones. Como el espacio de 12 dimensiones es dificil de visualizar utilizaremos la técnica PCA (Análisis de Componentes Principales) para analizar las relaciones entre características y así reducri la dimensionalidad.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

#Noramlizamos las caracteristicas para que estén en la misma escala
scaled_features = MinMaxScaler().fit_transform(features)

#Conseguir los dos principales componentes
pca = PCA(n_components=2).fit(scaled_features)
features_2d = pca.transform(scaled_features)
features_2d[0:10]  #Mostramos las 10 primeras instancias
```

```python
import matplotlib.pyplot as plt

plt.scatter(features_2d[:,0],features_2d[:,1])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Data')
plt.show()
```

Voy a utilizar una muestra de datos para crear una serie de modelos de agrupamiento con un número creciente de clusters y medir con qué precisión están agrupados los puntos de datos dentro de cada grupo. Para medir la compacidad de los grupos formados por el modelo usaremos la métrica WCSS.

Para determinar el número óptimo de clusters, utilizamos una técnica llamada "método del codo". Este método implica ajustar el modelo de agrupamiento con diferentes números de clusters y calcular el WCSS para cada modelo. Luego, se traza un gráfico del número de clusters en el eje x y el valor de WCSS en el eje y.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#CRear 10 modelos con 1 hasta 10 clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    # Fit the data points
    kmeans.fit(features.values)
    # Get the WCSS (inertia) value
    wcss.append(kmeans.inertia_)
    
#Plot the WCSS values onto a line graph
plt.plot(range(1, 11), wcss)
plt.title('WCSS by Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Podemos ver un cambio de pendiente notable para 2 y 4 clusters y un cambio de pendiente menos pronunciable para 3 y 5 clusters. El gráfico muestra una gran reducción en el WCSS de 1 a 2 clusters y también una reducción muy notable de dos a tres y de tres a cuatro. A aprtir de 4 clusters la reducción es menos notable. Podemos pensar que hay dos, tres o cuatro clusters de puntos bien separados.

## K-Means clustering

Utilizo el algoritmo K-Means para crear clusters de prueba. Este es un algoritmo de agrupamiento comúnmente utilizado que separa un conjunto de datos en K clusters de varianza igual. Probaré para K 4 ya que es cuando la curva más se aplana.

```python
from sklearn.cluster import KMeans

#Función para crear el modelo kmeans
def kmeans(n_clusters, data):
    model = KMeans(n_clusters=n_clusters, init='k-means++', n_init=100, max_iter=1000)
    clusters = model.fit_predict(data)
    return model, clusters

model, km_clusters = kmeans(4, features.values)
km_clusters
```

```python
import json

def load_team_names(file_path):
    with open(file_path, 'r') as file:
        team_names = json.load(file)
    return team_names

team_names = load_team_names('dataframe/id_equipo.json')
```

```python
#Función para dibujar los graficos con los clusters para los diferentes modelos
import matplotlib.pyplot as plt

local_labels = data['Local']

def plot_clusters(samples, clusters, team_names):
    unique_clusters = set(clusters)
    colors = plt.cm.tab10.colors[:len(unique_clusters)]  # Escoge una paleta de colores
    
    # Diccionario para almacenar los nombres de los clusters y sus colores correspondientes
    cluster_info = {}
    
    for cluster_id, color in zip(unique_clusters, colors):
        cluster_samples = samples[clusters == cluster_id]
        cluster_local_labels = local_labels[clusters == cluster_id]
        for sample, label in zip(cluster_samples, cluster_local_labels):
            plt.scatter(sample[0], sample[1], color=color, marker='o', s=20)
            plt.text(sample[0], sample[1] + 0.01, str(label), fontsize=8, ha='center', va='bottom')
        
        # Añadir una entrada al diccionario con el nombre del cluster y su color
        cluster_info[f'Cluster {cluster_id}'] = color
    
    # Crear leyenda personalizada
    legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=cluster_name,
                                 markerfacecolor=color, markersize=10) 
                      for cluster_name, color in cluster_info.items()]
    
    plt.legend(handles=legend_handles, loc='upper left', bbox_to_anchor=(1.05, 1))
    
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Assignments with Local Labels')
    plt.show()
```

```python
# Visualizar los clusters con los labels locales
plot_clusters(features_2d, km_clusters, team_names)
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

#Función para evaluar el modelo de clustering
def evaluate_clustering_model(data, labels):
    silhouette = silhouette_score(data, labels)
    calinski_harabasz = calinski_harabasz_score(data, labels)
    davies_bouldin = davies_bouldin_score(data, labels)
    
    print("Silhouette Score:", silhouette)
    print("Calinski-Harabasz Index:", calinski_harabasz)
    print("Davies-Bouldin Score:", davies_bouldin)
    
    return silhouette, calinski_harabasz, davies_bouldin
```

- Silhouette Score: Esta métrica cuantifica la cohesión y la separación de los clusters. Un valor cercano a 1 indica que los puntos están bien agrupados dentro de sus clusters y están separados de otros clusters. Un valor cercano a -1 indica que los puntos están mal asignados a sus clusters y podrían pertenecer a otros clusters. Un valor cercano a 0 indica una superposición de clusters. En general, valores más altos son mejores.

- Calinski-Harabasz Index: También conocido como criterio de Varianza entre grupos, es una medida de cuánto mejor son los clusters en comparación con los puntos distribuidos aleatoriamente. Valores más altos indican clusters más densos y bien separados.

- Davies-Bouldin Score: Esta métrica mide la compacidad dentro de los clusters y la separación entre los clusters. Un valor más bajo indica clusters mejor definidos.

```python
evaluate_clustering_model(features.values, km_clusters)
```

Los valores de las métricas de evaluación de clustering muestran un rendimiento aceptable. El Silhouette Score de 0.484 sugiere una separación razonable entre los clusters, donde valores más cercanos a 1 indican una mejor separación. El índice Calinski-Harabasz de 38.79 indica una densidad y separación adecuadas entre los clusters, mientras que el coeficiente de Davies-Bouldin de 0.806 sugiere una buena definición de los clusters con menos superposición entre ellos. En conjunto, estos resultados sugieren que el modelo de clustering podría estar proporcionando una agrupación coherente de los datos.

## Hierarchical Clustering


```python
from sklearn.cluster import AgglomerativeClustering

def agglomerative_clustering(n_clusters, data):
    agg_model = AgglomerativeClustering(n_clusters)
    agg_clusters = agg_model.fit_predict(data)
    return agg_clusters

agg_clusters = agglomerative_clustering(4, features.values)
agg_clusters
```

```python
plot_clusters(features_2d, agg_clusters, team_names)
```

```python
evaluate_clustering_model(features.values, agg_clusters)
```

```python
from sklearn.cluster import MeanShift

def find_optimal_bandwidth(features_2d, bandwidths):
    silhouette_scores = []
    
    for bandwidth in bandwidths:
        ms = MeanShift(bandwidth=bandwidth)
        ms.fit(features_2d)
        labels = ms.labels_
        silhouette_scores.append(silhouette_score(features_2d, labels))
    
    optimal_bandwidth = bandwidths[np.argmax(silhouette_scores)]
    return optimal_bandwidth

def mean_shift_clustering(features_2d, bandwidth):
    ms = MeanShift(bandwidth=bandwidth)
    ms.fit(features_2d)
    cluster_centers = ms.cluster_centers_
    cluster_labels = ms.labels_
    return cluster_centers, cluster_labels

bandwidths = np.arange(0.1, 1.0, 0.1)
optimal_bandwidth = find_optimal_bandwidth(features_2d, bandwidths)
cluster_centers, cluster_labels = mean_shift_clustering(features_2d, optimal_bandwidth)


plot_clusters(features_2d, cluster_labels, team_names)

evaluate_clustering_model(features_2d, cluster_labels)

```

Este modelo tiene valores más altos en todas las métricas de evaluación, lo que sugiere que produce clusters de mejor calidad en comparación con los otros modelos. Por lo tanto, basado en estas métricas de evaluación, este modelo es mejor.
